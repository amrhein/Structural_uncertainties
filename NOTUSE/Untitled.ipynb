```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(11)
test = np.cumsum(np.random.randn(300))

nc = .85
c = np.array([1,1,1])*nc

fig = plt.figure(figsize=(10,5))
plt.plot(test,color=c)
plt.axis('off')
plt.savefig('Figs/ts0.pdf')
plt.show();


fig = plt.figure(figsize=(10,5))
plt.plot(test,color=c)
N = 50
plt.plot(np.convolve(test,1./N*np.ones(N),'same'),'--',color=c)
N = 20
plt.plot(test,color=c)
plt.plot(np.convolve(test,1./N*np.ones(N),'same'),'--',color=c)
#plt.axis('off')
plt.savefig('Figs/ts50.pdf')
plt.show();


fig = plt.figure(figsize=(10,5))
N = 20
plt.plot(test,color=c)
plt.plot(np.convolve(test,1./N*np.ones(N),'same'),'--',color=c)
plt.axis('off')
plt.savefig('Figs/ts20.pdf')
plt.show();

fig = plt.figure(figsize=(10,5))
N = 10
plt.plot(test,color=c)
plt.plot(np.convolve(test,1./N*np.ones(N),'same'),'--',color=c)
plt.axis('off')
plt.savefig('Figs/ts10.pdf')
plt.show();

fig = plt.figure(figsize=(10,5))
N = 100
plt.plot(np.convolve(test,1./N*np.ones(N),'same'),color='C1')
plt.plot(test,color=c)
plt.axis('off')
plt.savefig('Figs/ts100.pdf')
plt.show();

```
